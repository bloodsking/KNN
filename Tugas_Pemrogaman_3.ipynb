```python
def euclideanA(data1, data2, data3, data4, data5, data6, i, index):
  dataD = []
  for j in range(index):
    rumus1 = ((data1[i] - data2[j])**2 + (data3[i] - data4[j])**2 + (data5[i] - data6[j])**2)**0.5
    dataD.append(((rumus1, j, x5[j])))

  return dataD
```
